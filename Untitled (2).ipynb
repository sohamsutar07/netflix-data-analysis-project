```python
import pandas as pd

df = pd.read_csv(r"D:\data analyst roadmap 1\project guide\netflix data analyst\netflix_p1.csv", encoding='latin1')

df.head()
```

```python
df['type'].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.show()
```

```python
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title("Yearly Content Addition")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top Countries")
plt.show()
```

```python
df['rating'].value_counts().head(5).plot(kind='pie')
plt.show()
```

```python
df['duration_number'] = pd.to_numeric(df['duration_number'], errors='coerce')
df['duration_number'].plot(kind='hist')
plt.title("Duration Distribution")
plt.show()
```
